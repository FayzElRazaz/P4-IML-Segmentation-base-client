### Espace de fonctions

```python
def suppr_outlier(df,col):
    q1 = round(np.percentile(df[col].dropna(),25),3)
    q3 = round(np.percentile(df[col].dropna(),75),3)
    bsup = q3+1.5*(q3-q1)
    binf = q1-1.5*(q3-q1)
    df.loc[df[col]>bsup,col]=np.nan
    df.loc[df[col]<binf,col]=np.nan
```

```python
def passage_au_log(df,col):
    return np.log(1+df.loc[:,col])
```

## RFM

```python
# Distribution des clients
dist = df[['customer_city', 'customer_unique_id']].\
                groupby(['customer_city']).\
                agg({'customer_unique_id': lambda x: x.nunique()}).\
                reset_index().sort_values('customer_unique_id',
                                          ascending=False)

dist.rename(columns={'customer_unique_id': 'nb clients'},
                    inplace=True)
dist.head()
```

```python
dist['nb clients'].sum()
```

```python
df.price.min()
```

```python
df.price.max()
```

```python
oldest = df["order_purchase_timestamp"].min()
oldest
```

```python
last = df["order_purchase_timestamp"].max()
last
```

```python
amj = str.split(str.split(last)[0], "-")
date_max = dt.datetime(int(amj[0]), int(amj[1]), int(amj[2]) + 1)
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
```

```python
RFM = df.groupby('customer_unique_id').\
            agg({"order_purchase_timestamp": lambda x: (date_max - x.max()).days,
                 "order_id": lambda x: x.nunique(),
                 "price": lambda x: sum(x)})
```

```python
RFM['order_purchase_timestamp'] = RFM['order_purchase_timestamp'].astype(int)
```

```python
RFM.rename(columns={'order_purchase_timestamp': 'recency',
                         'order_id': 'frequency',
                         'price': 'monetary_value'}, inplace=True)
```

```python
RFM.describe()
```

```python
df_final = df.merge(RFM,left_on = "customer_unique_id",right_on = "customer_unique_id")
```

```python
df_final.head()
```

```python
df_final.drop(["customer_id","order_id","order_status","order_approved_at","order_delivered_carrier_date",\
              "order_delivered_customer_date","order_estimated_delivery_date","order_item_id","product_id",\
              "seller_id","shipping_limit_date","review_id","review_comment_title","review_comment_message",\
              "review_creation_date","review_answer_timestamp"], axis = 1,inplace = True)
```

```python
df_final.head()
```

```python
tup_int64 = []
tup_obj = []
tup_float64 = []
for i in range(len(df_final.columns)):
    if(df_final[df_final.columns[i]].dtype == 'int64'):
        tup_int64.append(df_final.columns[i])
    elif(df_final[df_final.columns[i]].dtype == 'float64'):
        tup_float64.append(df_final.columns[i])
    else:
        tup_obj.append(df_final.columns[i])
```
